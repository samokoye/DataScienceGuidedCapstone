This project will aim at figuring out how much an investor would have made if he invested $1000 in some selected stocks on January 1st 2010. This project will track the investor's 10 year stock history and determine how much returns he/she currently have. It will find possible losses and which sector of the investment did better than the other. The term "did better" will be categorized into two. First is an overall performance without considering COVID pandemic, and the second will compare each stock against one another to see which "did better" during the pandemic. The sectors to be analyzed in this analysis are Manufacturing & Materials(36 percent), Consumer Goods(23 percent), Energy & Water(17 percent), Healthcare(11 percent), Business(6 percent), and Technology(7 percent). 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import requests
import yfinance as yf
import sqlalchemy
from functools import reduce
```
